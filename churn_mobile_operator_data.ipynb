```python
data['DSL'] = new_data['DSL']
data['Fiber optic'] = new_data['Fiber optic']
data['No'] = new_data['No']
data.drop(['Internet Service'] , axis='columns' , inplace=True)
```

```python
new_data = pd.get_dummies(data['Payment Method'])
```

```python
data['Bank transfer'] = new_data['Bank transfer (automatic)']
data['Credit card'] = new_data['Credit card (automatic)']
data['Electronic check'] = new_data['Electronic check']
data['Mailed check'] = new_data['Mailed check']

data.drop(['Payment Method'] , axis='columns' , inplace=True)
```

```python
data.head()
```

```python
data['Churn'] = data['Churn'].apply(lambda x:1 if x=='Yes' else 0)
```

```python
data.head()
```

**CONVERTING TENURE COLUMN IN SCALER FORMAT FOR BETTER RESULT**

```python
data.tenure.max()
```

```python
data['tenure'] = data.tenure/72
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Total Charges'] = pd.to_numeric(data['Total Charges'] , errors = 'coerce')
data['Contract'] = pd.to_numeric(data['Contract'] , errors='coerce')
```

```python
data.dtypes
```

```python
X = data.drop(['Churn'] , axis='columns')
```

```python
y = data.Churn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
x_train
```

```python
x_train.shape
```

```python
model = keras.Sequential([
        keras.layers.Dense(20 , input_shape = (25,) , activation='relu'),
        keras.layers.Dense(15 , activation='relu'),
        keras.layers.Dense(10 ,  activation='relu'),
        keras.layers.Dense(1 , activation='sigmoid'),
])

model.compile(
    optimizer = 'SGD',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train , y_train, epochs=5)
```

```python
model.evaluate(x_test , y_test)
```
